```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals import joblib

efe=pd.read_csv('Best Movie by Year Netflix.csv')
X=efe.drop(columns=['TITLE'])
y=efe['TITLE']

model=DecisionTreeClassifier()
model.fit(X,y)

joblib.dump(model,'filmgpt.joblib')
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals import joblib

model=joblib.load('filmgpt.joblib')
predictions=model.predict(['drama,US'])
predictions
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

efe=pd.read_csv('Best Movie by Year Netflix.csv')
X=efe.drop(columns=['TITLE'])
y=efe['TITLE']

model=DecisionTreeClassifier()
model.fit(X,y)


tree.export_graphviz(model, out_file='filmgpt.dot',
                    feature_names=['title','release_year'],
                    class_names=sorted(y.unique()),
                     label='all',
                     rounded=True,
                     filled=True)
```
